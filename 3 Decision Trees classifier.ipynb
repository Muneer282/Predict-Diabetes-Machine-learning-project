# Decision Trees classifiers

# 1.Modeling

```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
import seaborn as sns 
```

```python
my_data = pd.read_csv("cleanedData.csv")
```

```python
y = my_data['Outcome']
X = my_data.drop(['Outcome'], axis=1)
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.1, random_state=32)
```

```python
DiabetesTree = DecisionTreeClassifier(criterion="entropy", max_depth=4)
DiabetesTree.fit(X_trainset, y_trainset)
```

### 2.Evaluation


```python
pred = DiabetesTree.predict(X_testset)
print("Decision Trees's Accuracy: ", metrics.accuracy_score(y_testset, pred))
```
